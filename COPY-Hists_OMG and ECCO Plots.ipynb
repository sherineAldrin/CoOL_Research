```python
# returns distances between my point and all the other points, from Prof
def great_circle_distance(lon_ref, lat_ref, Lon, Lat):
    earth_radius = 6371000
    lon_ref_radians = np.radians(lon_ref)
    lat_ref_radians = np.radians(lat_ref)
    lons_radians = np.radians(Lon)
    lats_radians = np.radians(Lat)
    lat_diff = lats_radians - lat_ref_radians
    lon_diff = lons_radians - lon_ref_radians
    d = np.sin(lat_diff * 0.5) ** 2 + np.cos(lat_ref_radians) * np.cos(lats_radians) * np.sin(lon_diff * 0.5) ** 2
    h = 2 * earth_radius * np.arcsin(np.sqrt(d))
    return(h)
```

### make plots too

```python
import csv
import pandas as pd
```

```python
ecco_files = []
ecco_dir = "/Users/sherine_aldrin/Downloads/CoOL/GreenlandModel/"

# Define which months exist for each year based on OMG data availability
available_months = {
    "2016": ["9", "10"],   # Sept and Oct
    "2017": ["10"],        # Only Oct
}

for variable in ["THETA", "SALT"]:
    for year in ["2016", "2017"]:
        ecco_path = f"{ecco_dir}{variable}_{year}.nc"
        if not os.path.exists(ecco_path):
            print(f"Missing ECCO file: {ecco_path}")
            continue

        ds_ecco = xr.open_dataset(ecco_path)

        # Only include months for which OMG data exists
        for month in available_months[year]:
            ecco_files.append((ds_ecco, variable, year, month, None))
```

```python

# Organizing and plotting all the ecco data now

lon = ds_grid['XC']
lat = ds_grid['YC']
depth_ecco = Z.values
#depth_mask = ds_grid['Depth'].values
#hFacC = ds_grid['hFacC'].values  # for masking just in case there's land

# Folder to save plots
plot_folder = "/Users/sherine_aldrin/Downloads/CoOL/GreenlandModel/NEW_Files_OMG_vs_ECCO"
os.makedirs(plot_folder, exist_ok=True)

def get_all_nc_files(folder):
    nc_files = []
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.endswith(".nc"):
                nc_files.append(os.path.join(root, file))
    return nc_files
```

```python
summary_rows = []

for ds_ecco, variable, year, month, omg_files in ecco_files:
    month_index = int(month) - 1
    ecco_data = ds_ecco[variable].isel(time=month_index)
    depth_ecco_full = Z.values * -1  # flip ECCO depth for plotting

    lon_ecco = lon.values
    lat_ecco = lat.values

    # Flatten grid for distance calculations
    flat_lon = lon.values.reshape(-1)
    flat_lat = lat.values.reshape(-1)
    flat_indices = np.arange(flat_lon.size)

    # Adjust folder name
    month_folder_name = "09_September" if month == "9" else "10_October"
    omg_folder = f"/Users/sherine_aldrin/Downloads/CoOL/GreenlandModel/Monthly_OMG_{year}/{month_folder_name}"
    omg_files = get_all_nc_files(omg_folder)

    for omg_file in omg_files:
        omg_ds = xr.open_dataset(omg_file, engine="netcdf4")

        lat_omg = omg_ds.attrs["latitude"]
        lon_omg = omg_ds.attrs["longitude"]

        dists = great_circle_distance(lon_omg, lat_omg, flat_lon, flat_lat)
        
        #min_idx = np.argmin(dists)
        #flat_index = flat_indices[min_idx]
        #tile, j, i = np.unravel_index(flat_index, lon_ecco.shape)
        #ecco_profile = ecco_data[:, tile, j, i].values

        # Sort indices by distance (closest first)
        sorted_indices = flat_indices[np.argsort(dists)]

        # Search for the closest ECCO point with valid data
        valid_point_found = False
        for flat_index in sorted_indices:
            tile, j, i = np.unravel_index(flat_index, lon_ecco.shape)
            profile = ecco_data[:, tile, j, i].values

            if np.any(~np.isnan(profile)) and not np.all(profile == 0):
                ecco_profile = profile
                valid_point_found = True
                break  # stop at the first valid one

        if not valid_point_found:
            print(f"WARNING: No valid ECCO profile found near {omg_file} — skipping.")
            continue
        
        # Trim ECCO profile to ignore trailing zeros
        nonzero = np.nonzero(ecco_profile)[0]
        if nonzero.size > 0:
            last_valid = nonzero[-1] + 1
            ecco_trimmed = ecco_profile[:last_valid]
            depth_trimmed = depth_ecco_full[:last_valid]
        else:
            ecco_trimmed = np.array([])
            depth_trimmed = np.array([])

        # Use correct OMG variable based on ECCO variable
        
        if variable == "THETA":
            omg_profile = omg_ds["potential_temperature"].values
        elif variable == "SALT":
            omg_profile = omg_ds["practical_salinity"].values
        else:
            print(f"Unsupported variable: {variable}")
            continue

        depth_omg = omg_ds["depth"].values * -1


        # GET AVERAGES
        ecco_mean = np.nanmean(ecco_trimmed[ecco_trimmed != 0]) if ecco_trimmed.size > 0 else np.nan
        omg_mean = np.nanmean(omg_profile)
        diff = omg_mean - ecco_mean

        # Save summary data
        summary_rows.append({
            "file": os.path.basename(omg_file),
            "variable": variable,
            "latitude": lat_omg,
            "longitude": lon_omg,
            "ecco_mean": ecco_mean,
            "omg_mean": omg_mean,
            "difference": diff
        })
        
        fig = plt.figure()
        

        plt.plot(omg_profile, depth_omg, label="OMG", color="blue")
        plt.plot(ecco_profile, depth_ecco, label="ECCO", color="red")

        # the dotted lines
        plt.axvline(ecco_mean, color="red", linestyle=":", label=f"ECCO Mean: {ecco_mean:.2f}")
        plt.axvline(omg_mean, color="blue", linestyle=":", label=f"OMG Mean: {omg_mean:.2f}")
        
        plt.ylim(depth_omg.min(), depth_omg.max())  # Limit to OMG depth
        #plt.gca().invert_yaxis()

        # Label based on variable
        if variable == "THETA":
            plt.xlim(-10, 15)
            plt.xlabel("Temperature (°C)")
        elif variable == "SALT":
            plt.xlim(20, 40)
            plt.xlabel("Salinity (psu)")
        else:
            plt.xlabel(variable)

        plt.ylabel("Depth (m)")
        plt.title(f"{variable} Profile (ECCO vs OMG) {year}, {month}")
        plt.legend()
        
        # Save plot
        base_name = os.path.splitext(os.path.basename(omg_file))[0]
        month_folder_out = os.path.join(plot_folder, f"{year}_{month}")
        os.makedirs(month_folder_out, exist_ok=True)
        plot_path = os.path.join(
            month_folder_out, f"{variable}_{base_name}_vs_ECCO_{year}_{month}.png"
        )
        plt.tight_layout()
        plt.savefig(plot_path, dpi=300)
        plt.close(fig)

    csv_path = os.path.join(plot_folder, f"profile_summary_{variable}_{year}_{month}.csv")
    with open(csv_path, "w", newline="") as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=["file", "variable", "latitude", "longitude", "ecco_mean", "omg_mean", "difference"])
        writer.writeheader()
        writer.writerows(summary_rows)

summary_df = pd.DataFrame(summary_rows)

# Plot histogram of differences
for var in summary_df["variable"].unique():
    var_df = summary_df[summary_df["variable"] == var]
    diffs = var_df["difference"].dropna().values
    mean_diff = np.mean(diffs)

    plt.figure()
    plt.hist(diffs, bins=20, color="blue", edgecolor="black")
    plt.axvline(mean_diff, color="red", linestyle="--", label=f"Mean Diff = {mean_diff:.2f}")
    plt.xlabel(f"{var} Difference (OMG - ECCO)")
    plt.ylabel("Count")
    plt.title(f"Distribution of Differences ({var})")
    plt.legend()
    plt.grid(True)

    # Save histogram
    hist_path = os.path.join(plot_folder, f"{var}_difference_histogram.png")
    plt.tight_layout()
    plt.savefig(hist_path, dpi=300)
    plt.close()
```

### Make the sorted histograms

```python
# Load summary difference data and region info
summary_csv = "/Users/sherine_aldrin/Downloads/CoOL/GreenlandModel/NEW_Files_OMG_vs_ECCO"
all_csv_paths = glob.glob(os.path.join(summary_csv, "profile_summary_*.csv"))
region_csv = "/Users/sherine_aldrin/Downloads/CoOL/GreenlandModel/OMG_CTD_Locations_2016_2017_with_region.csv"

# Combine all CSVs
summary_df_list = [pd.read_csv(csv) for csv in all_csv_paths]
summary_df = pd.concat(summary_df_list, ignore_index=True)

region_csv = "/Users/sherine_aldrin/Downloads/CoOL/GreenlandModel/OMG_CTD_Locations_2016_2017_with_region.csv"
region_df = pd.read_csv(region_csv)

summary_df["file_clean"] = summary_df["file"].str.replace("CTD_", "", regex=False).str.replace(".nc", "", regex=False)
region_df["File_ID"] = region_df["File_ID"].str.strip()

merged_df = pd.merge(
    summary_df,
    region_df[["File_ID", "Region"]].rename(columns={"File_ID": "file_clean"}),
    on="file_clean",
    how="left"
)
print("Matched profiles with region info:", merged_df["Region"].notna().sum())

# Drop rows with missing region or difference
merged_df = merged_df.dropna(subset=["Region", "difference"])

# Regions and their positions in GridSpec
region_positions = {
    "N": (0, 0),
    "NW":  (1, 0),
    "CW": (2, 0),
    "SW": (3, 0),
    "NE": (0, 1),
    "CE": (1, 1),
    "SE": (2, 1)
}
regions = list(region_positions.keys())

# Output folder
plot_folder = "/Users/sherine_aldrin/Downloads/CoOL/GreenlandModel/Files_OMG_vs_ECCO/region_histograms"
os.makedirs(plot_folder, exist_ok=True)

# Loop over variables and make GridSpec histogram plots
for variable in ["THETA", "SALT"]:
    fig = plt.figure(figsize=(14, 16))
    gs = gridspec.GridSpec(4, 2, figure=fig)
    fig.suptitle(f"OMG - ECCO {variable} Differences by Region", fontsize=16)

    var_df = merged_df[merged_df["variable"] == variable]

    for region in regions:
        row, col = region_positions[region]
        ax = fig.add_subplot(gs[row, col])

        region_data = var_df[var_df["Region"] == region]["difference"].dropna().values
        region_data = region_data[np.abs(region_data) < 5]  # Remove outliers

        if len(region_data) == 0:
            ax.set_title(f"{region} (No Data)")
            ax.axis("off")
            continue

        mean_diff = np.mean(region_data)
        ax.hist(region_data, bins=30, color="blue", edgecolor="black")
        ax.axvline(mean_diff, color="red", linestyle="--", label=f"Mean = {mean_diff:.2f}")
        ax.set_title(region)
        ax.set_xlabel("OMG - ECCO")
        ax.set_ylabel("Count")
        ax.grid(True)
        ax.legend()

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    fname = f"{variable}_region_histograms.png".replace(" ", "_")
    plt.savefig(os.path.join(plot_folder, fname), dpi=300)
    plt.close()

print(f"Saved GridSpec region histograms to: {plot_folder}")
```
